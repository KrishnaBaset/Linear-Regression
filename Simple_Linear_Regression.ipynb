```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
```

```python
dt=pd.read_csv('C:\\Users\\Krishnakant Baset\\OneDrive\\Desktop\\Data Science Material\\delivery_time.csv')
dt.head()
```

```python
dt.info()
```

```python
dt.corr()
```

```python
sns.set_style(style='darkgrid')
sns.pairplot(dt)
```

**Preparing a model**

```python
import statsmodels.formula.api as smf
model = smf.ols("Delivery_Time~Sorting_Time",data = dt).fit()
sns.regplot(x="Delivery_Time", y="Sorting_Time", data=dt);
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)  
```

```python
(model.rsquared,model.rsquared_adj)
```

**Predict For Data**

```python
#Predict for 5 and 8 Sorting Time
newdata=pd.Series([5,8])
```

```python
data_pred=pd.DataFrame(newdata,columns=['Sorting_Time'])
model.predict(data_pred)
```

**Model 2 - Salary Hike**

```python
salarydt=pd.read_csv('C:\\Users\\Krishnakant Baset\\OneDrive\\Desktop\\Data Science Material\\Salary_Data.csv')
salarydt.head()
```

```python
salarydt.info()
```

```python
salarydt.corr()
```

```python
sns.set_style(style='darkgrid')
sns.pairplot(salarydt)
```

```python
sns.regplot(x="YearsExperience", y="Salary", data=salarydt);
```

**Preparing a Model**

```python
import statsmodels.formula.api as smf
salarymodel = smf.ols("Salary~YearsExperience",data = salarydt).fit()
print(salarymodel.tvalues, '\n', salarymodel.pvalues) 
```

```python
(salarymodel.rsquared,salarymodel.rsquared_adj)
```

**Predict for data**

```python
#Predict salary for 1.7 and 2.1 YearsExperience
newdata1=pd.Series([1.7,2.1])
```

```python
data_pred1=pd.DataFrame(newdata1,columns=['YearsExperience'])
salarymodel.predict(data_pred1)
```
